````python
# Create new dataframe
state_region = list()
state = ""

# Using with open, to open textfile
with open('university_towns.txt') as f: # This closes file for us
    for line in f:
        if '[edit]' in line:
            state = line[:line.find('[edit]')]
        
        elif "(" in line:
            region = line[:(line.find('(') - 1)]
            state_region.append([state,region])
        elif "\n" in line:
            region = line[:(line.find('\n'))]
            state_region.append([state,region])
        else:
            region = line
            state_region.append([state,region])
    df = pd.DataFrame(state_region, columns =['State', 'RegionName'])




state_region
````

````python
# Load excel file
gdp = pd.read_excel('gdplev.xls')

# Useless columns
useless = [0,1,2,3,7]

# Drop useless columns
gdp.drop(gdp.columns[useless], axis =1, inplace= True)
gdp = gdp[7:]
gdp.columns = ["Quarterly","GDP in billions of current dollars","GDP in billions of chained 2009 dollars"]

# Subset only year 2000 and quart 1 and after
# Find index of 2000q1
getindex = gdp.loc[gdp['Quarterly']== '2000q1'].index.values.item()
getindex

# Subset gdp dataframe
gdp = gdp.ix[getindex:]

gdp
````

````python
gdp.reset_index(inplace=True)
````

````python
# Find Recession Start
start = []

for i in range(0, len(gdp)-1):
    if (gdp.iloc[i]['GDP in billions of chained 2009 dollars'] < gdp.iloc[i-1]['GDP in billions of chained 2009 dollars']) and (gdp.iloc[i+1]['GDP in billions of chained 2009 dollars'] < gdp.iloc[i]['GDP in billions of chained 2009 dollars']) :
        start.append(gdp.iloc[i])

recession = start[0]['Quarterly']

recession
````

````python
def get_recession_start():
    '''Returns the year and quarter of the recession start time as a 
    string value in a format such as 2005q3'''
    
    return recession
````

````python
# Get index of 2008q3

startindex = gdp.loc[gdp['Quarterly']== recession].index.values.item()
last= []

for i in range(startindex+2, len(gdp)):
    if (gdp.iloc[i]['GDP in billions of chained 2009 dollars'] >  gdp.iloc[i-1]['GDP in billions of chained 2009 dollars']) and (gdp.iloc[i-1]['GDP in billions of chained 2009 dollars'] > gdp.iloc[i-2]['GDP in billions of chained 2009 dollars']):
        last.append(gdp.iloc[i])
        
endrecession = last[0]['Quarterly']
endrecession

````

````python
def get_recession_end():
    '''Returns the year and quarter of the recession end time as a 
    string value in a format such as 2005q3'''
       
    return endrecession
````

````python
# Find end index
endindex = gdp.loc[gdp['Quarterly'] == endrecession].index.values.item()

# Dataframe with recession period
recessionper = gdp.iloc[startindex:endindex]
filter = recessionper['GDP in billions of chained 2009 dollars'] == recessionper['GDP in billions of chained 2009 dollars'].min()
bottom = recessionper.loc[filter]
bot = bottom['Quarterly'].item()
````

````python
def get_recession_bottom():
    '''Returns the year and quarter of the recession bottom time as a 
    string value in a format such as 2005q3'''
    
    return bot
````

````python
# Read in housing data
housing = pd.read_csv('City_Zhvi_AllHomes.csv')

# Drop Useless Columns
housing.drop(housing.columns[[0,3,4,5,6]], axis =1 , inplace= True)

housing['State'] = housing['State'].map(states)
housing = housing.set_index(['State', 'RegionName']).convert_objects(convert_numeric=True)


# Find Index of 2001-01
index = list(housing.columns.values).index('2000-01')

#Remove more useless columns
housing.drop(housing.columns[2:index], axis = 1, inplace = True)

#Converting
for i in range(2, len(housing.columns)):
    columns = housing.iloc[:,i].name
    year = int(columns.split('-')[0])
    month = int(columns.split('-')[1])
    if month <= 3:
        quarter = 1
    elif month <= 6:
        quarter = 2
    elif month <= 9:
        quarter = 3
    elif month <= 12:
        quarter = 4
    quarter = str(year) + 'q' + str(quarter)
    if i:
        split = housing.iloc[:, i-2:i+1]  
    else:
        split = housing.iloc[:, i:len(housing.columns)]
    housing[quarter] = split.mean(axis=1)
    
housing.drop(housing.columns[:202], axis =1, inplace= True)

  
housing
````

````python


def convert_housing_data_to_quarters():
    '''Converts the housing data to quarters and returns it as mean 
    values in a dataframe. This dataframe should be a dataframe with
    columns for 2000q1 through 2016q3, and should have a multi-index
    in the shape of ["State","RegionName"].
    
    Note: Quarters are defined in the assignment description, they are
    not arbitrary three month periods.
    
    The resulting dataframe should have 67 columns, and 10,730 rows.
    '''
    
    return housing
````

````python
# Make new dataframes

univ_list = get_list_of_university_towns()
start = get_recession_start()
bottom = get_recession_bottom()
ending = get_recession_end()
quarters = convert_housing_data_to_quarters()

data = quarters.loc[:,'2008q3':'2009q4'].reset_index()

data['ratio'] = (data['2008q3']- data['2009q4'])/ data['2008q4']

# get list of towns

univ_list = univ_list['RegionName'].tolist()

data['UniTown'] = data.RegionName.apply(lambda x: x in univ_list)

uni = data[data.UniTown].copy().dropna()
not_uni= data[~data.UniTown].copy().dropna()

p = ttest_ind(uni['ratio'], not_uni['ratio'])[1]
different = p < 0.01 

lower = 'university town' if uni['ratio'].mean() < not_uni['ratio'].mean() else 'non-university town'

````

````python
def run_ttest():
    '''First creates new data showing the decline or growth of housing prices
    between the recession start and the recession bottom. Then runs a ttest
    comparing the university town values to the non-university towns values, 
    return whether the alternative hypothesis (that the two groups are the same)
    is true or not as well as the p-value of the confidence. 
    
    Return the tuple (different, p, better) where different=True if the t-test is
    True at a p<0.01 (we reject the null hypothesis), or different=False if 
    otherwise (we cannot reject the null hypothesis). The variable p should
    be equal to the exact p value returned from scipy.stats.ttest_ind(). The
    value for better should be either "university town" or "non-university town"
    depending on which has a lower mean price ratio (which is equivilent to a
    reduced market loss).'''
    

    return (different, p, lower)

run_ttest()
````
